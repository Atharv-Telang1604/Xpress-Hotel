```python
import json
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
with open("/content/data_set.json", "r") as file:
    hotel_data = json.load(file)
```

```python
rows = []

for day in hotel_data["data"]:
    for sale in day["sales"]:
        rows.append({
            "amount": sale["amount"],
            "payment_mode": sale["payment_mode"],
            "room_type": sale["room_type"],
            "is_cancelled": int(sale["is_cancelled"])
        })

df = pd.DataFrame(rows)
```

```python
X = pd.get_dummies(df[["amount", "payment_mode", "room_type"]])
y = df["is_cancelled"]
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
print("\nEnter booking details for cancellation risk prediction:")

amount = int(input("Enter booking amount (₹): "))
payment_mode = input("Enter payment mode (Cash / Card / UPI): ").strip()
room_type = input("Enter room type (Standard / Deluxe / Suite): ").strip()

sample_booking = {
    "amount": amount,
    "payment_mode": payment_mode,
    "room_type": room_type
}

sample_df = pd.DataFrame([sample_booking])
sample_encoded = pd.get_dummies(sample_df)
sample_encoded = sample_encoded.reindex(columns=X.columns, fill_value=0)

probability = model.predict_proba(sample_encoded)[0][1]

print(" ML-Based Cancellation Risk Prediction\n")
print(" Input Booking Details:")
print(f"• Payment Mode : {sample_booking['payment_mode']}")
print(f"• Room Type    : {sample_booking['room_type']}")
print(f"• Amount       : ₹{sample_booking['amount']}")

print("\n Model Used:")
print("• Logistic Regression")

print(f"\n Predicted Cancellation Probability: {probability:.2f}")

if probability > 0.6:
    print("Risk Level: HIGH")
elif probability > 0.3:
    print(" Risk Level: MEDIUM")
else:
    print(" Risk Level: LOW")

print("\n Note: Prediction is advisory due to limited historical data.")
```
